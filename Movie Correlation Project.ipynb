```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) # This Adjusts the configuration of the plots we will create


# Read in the data
df = pd.read_csv(r'C:\Users\Lionel\Downloads\movies.csv')
```

```python
df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# what are the data types for our columns?

df.dtypes
```

```python
# Change dtype of columns


df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create Correct Year Column

df['yearcorrect'] = df['released'].astype(str).str[:4]

df.head()
```

```python
df1 = df.sort_values(by=['gross'], inplace=False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop Duplicates

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
# Prediction: budget, company has high correlation to revenue
```

```python
# visualize prediction:

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget For Film')

plt.show()
```

```python
# for interpretation purposes df1 was assigned to the data that was sorted in ranking of gross revenue in descending order:

df1.head()
```

```python
# Plot the Budget vs Gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df1, scatter_kws = {"color": "red"}, line_kws = {"color": "blue"})
```

```python
#  Let's start looking at correlation: (Always be aware of the correlation method you're using because results are affected)

df1.corr(method = 'pearson') # corr types: pearson, kendall, spearman
```

```python
#  there's a high correlation between budget and gross
# prediction cofirmed true
```

```python
correlation_matrix = df1.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Look at Company
df1.head()
```

```python
df_numerized = df1

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
         df_numerized[col_name] = df_numerized[col_name].astype('category')
         df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df1.head()
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr.head()
```
